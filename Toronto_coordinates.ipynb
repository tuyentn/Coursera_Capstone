# Get Torono coodinates

<b>Import packages</b>

```python
import geocoder
import pandas as pd
import numpy as np
```

<b>Load postal code</b>

```python
postal_df = pd.read_csv(r'data\toronto_portal.csv', index_col=0)
postal_df.head()
```

<b>Get coordinates for postal</b>

```python
for postal in postal_df:
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
      g = geocoder.google('{}, Toronto, Ontario'.format(postal))
      lat_lng_coords = g.latlng

    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
```

*Use csv data*

```python
coordinates_df = pd.read_csv(r'data\Geospatial_Coordinates.csv')
coordinates_df.head()
```

<b>Join to postal</b>

```python
toronto_df = postal_df.join(coordinates_df.set_index('Postal Code'), on='PostalCode')
```
